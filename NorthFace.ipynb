```python
def part_of_speech_extraction(text):
    doc = nlp(text)
    list_Verbs_NS = [token.lemma_ for token in doc if (token.pos_ == "VERB" and token.is_stop == False)]
    list_Noun_NS = [token.lemma_ for token in doc if (token.pos_ == "NOUN" and token.is_stop == False)]
    list_Adj_NS = [token.lemma_ for token in doc if (token.pos_ == "ADJ" and token.is_stop == False)]
    return(list_Verbs_NS,list_Noun_NS, list_Adj_NS)
```

```python
df[["Verbs_NS", 
   "Noun_NS",
   "Adj_NS"]] = df.apply(lambda row: part_of_speech_extraction(row["sub_1"]), # Usage d'une fonction auxiliaire. 
                         axis = 1,
                         result_type='expand') # Pour associer le resulat en position i Ã  la colonne i. 
```

# Trouver le produit le + proche d'un autre. 
## IndÃ©pendant du K-Means. On se place aprÃ¨s le prÃ©processing. 
## Principe:
- Calculer la cosine_similarity entre nos vecteurs et les autres. 
- On prends celui avec la distance la + faible. 

```python
vect1 = df_after_preprocessing.iloc[5]
vect1
```

```python
vect2 = df_after_preprocessing.iloc[25]
vect2
```

```python
from scipy import spatial
similarity = 1 - spatial.distance.cosine(vect1, vect2)
similarity # SimilaritÃ© entre vect1 (ligne 5) et vect2 (ligne 25)
```

```python
def sorted_closer_vector(vect, df):
    df2 = df.copy()
    cosine_similarity = df2.apply(lambda row: 1 - spatial.distance.cosine(vect, row), axis = 1)
    #print(cosine_similarity)
    #print(cosine_similarity.sort_values(ascending = False))
    #list_cosine_similarity = [1 - spatial.distance.cosine(vect, elt) for elt in list_vect]
    return(cosine_similarity.sort_values(ascending = False)) #list(cosine_similarity.sort_values(ascending = False).index)) #(argmax(list_cosine_similarity))
    
```

```python
sorted_closer_vector(vect1, df_after_preprocessing)
```

```python
def found_closer_vector(id_product, df_vector, original_df, skip_best_cosine = True ):
    vector_product= df_vector.loc[id_product]
    sorted_cosine = sorted_closer_vector(vector_product, df_vector)
    print(sorted_cosine)
    print("Notre produit original est le {}".format(id_product))
    print("Sa description est {}".format(original_df.loc[id_product]["description"]))
    print('\n ####### \n')
    if skip_best_cosine: 
        pos_to_find = 1
    else:
        pos_to_find = 0
    id_product_found = sorted_cosine.index[pos_to_find]
    similarity = round(100 * sorted_cosine.iloc[pos_to_find], 1 )
    print(id_product_found, similarity)
    print("Le produit le plus est le NÂ°{}, avec une similaritÃ© de {}%".format(id_product_found, similarity))
    print("Sa description est {}".format(original_df.loc[id_product_found]["description"]))
    return()
```
